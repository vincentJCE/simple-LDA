```python
#生成一个数据集，属性空间是二维欧式空间，标记是{0,1}，用于LDA
#考虑假如噪声。
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
X1 = np.random.randn(100,2)
X2 = np.random.randn(100)
Y1 = np.ones(100)
Y2 = np.zeros(100)
X1 = X1 + (1,1)
save = pd.DataFrame({'pos':Y1,'label':X2})
save.to_csv('d:\LDA_DATA.csv')

X1 = np.random.randn(100) + 1
Y1 = np.random.randn(100) + 1
Label1 = np.ones(100)
X2 = np.random.randn(100)
Y2 = np.random.randn(100) 
Label2 = np.zeros(100)
X = np.append(X1, X2)
Y = np.append(Y1,Y2)

Label = np.append(Label1, Label2)
save = pd.DataFrame({'a':X,'b':Y,'c':Label})
save.to_csv('d:\LDA_DATA.csv')
```
